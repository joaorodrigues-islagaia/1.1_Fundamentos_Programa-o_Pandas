## 2.1 Estruturas de Dados Fundamentais

O problema central que o pandas se props a resolver foi o do alinhamento de dados intrnseco. Em ferramentas baseadas em arrays puros, como o NumPy a referncia aos dados  estritamente posicional. No mundo real, e especificamente em finanas, dados raramente chegam perfeitamente alinhados ou limpos. Uma srie temporal de preos de aes de uma empresa pode no ter registos nos mesmos dias que outra devido a feriados locais ou falhas de coleta. O pandas introduziu a filosofia de que os dados devem ser carregar com os seus metadados (o ndice), permitindo que operaes binrias respeitem os rtulos em vez da posio fsica na memria, garantindo a integridade da anlise mesmo na presena de dados ausentes ou dessincronizados.

Para compreender "como funciona" o pandas,  necessrio analisar para l da sintaxe e examinar a engenharia (arquitetura) subjacente que permite as suas operaes. O pandas no  uma ferramenta isolada; ele  construdo sobre a base do NumPy (como j referido anteriormente), mas estende suas capacidades atravs de uma camada de abstrao sofisticada que gerencia a complexidade dos tipos de dados mistos, enquanto no NumPy, a estrutura de dados primria  o ndarray (N-dimensional array), que exige homogeneidade de tipos para garantir performance.

O pandas resolve essa incompatibilidade entre a necessidade de heterogeneidade dos dados do mundo real e a exigncia de homogeneidade do NumPy atravs de um componente interno denominado **BlockManager**. O BlockManager atua como um orquestrador de memria. Quando um usurio cria um DataFrame com mltiplas colunas de tipos diferentes, o BlockManager no aloca cada coluna separadamente na memria, nem tenta forar tudo em um nico array de tipo genrico (o que seria lento). Em vez disso, ele agrupa colunas de tipos idnticos em "blocos".

Por exemplo, se um DataFrame possui dez colunas, sendo trs de nmeros inteiros, cinco de valores decimais  e duas de texto, o BlockManager manter, em teoria, trs blocos de memria distintos: uma matriz NumPy 3xN para os inteiros, uma 5xN para os floats e um array de objetos para as strings. Essa arquitetura permite que o pandas delegue operaes matemticas para o NumPy dentro de cada bloco homogneo, mantendo a velocidade de execuo.

A arquitetura terica do Pandas assenta em dois tipos de objetos primrios: a Series e o DataFrame. A compreenso da distino e da relao entre estes dois objetos  conceptualmente importante.

Outro elemento estrutural com elevada importncia, e potencial, so os Indices.


### 2.1.1 _ndice_

O ndice  o que separa funcionalmente o Pandas do NumPy. Pandas usa uma tcnica chamada tabela hash para guardar os ndices. O Indice representa a capacidade de **"lookup"** (procurar, como usamos por exemplo no Excel). No conjunto de dados do Titanic, o ndice padro ser um **RangeIndex** (0, 1, 2...), mas teoricamente poderamos definir o ndice como sendo o *PassengerId* ou at mesmo o *Name*. Isto permite a chamada de dados baseada em indice significativos em vez de posies arbitrrias, aproximando a manipulao de dados da lgica de bases de dados relacionais (SQL). 

 como um endereo ou etiqueta para cada linha da tabela. Em vez de procurar os dados linha por linha, o Pandas usa este ndice para encontrar rapidamente o que precisa. 
Exemplo: Se tens uma lista de nomes, o ndice pode ser os nmeros (0, 1, 2) ou os prprios nomes.  

Existem 3 tipos de ndices que respondem a lgicas distintas e especficas: 

- **RangeIndex**: Uma representao otimizada de uma sequncia de inteiros (0, 1, 2...), semelhante  funo range do Python, que consome memria insignificante independentemente do tamanho do DataFrame.

- **DatetimeIndex**: Talvez a contribuio mais significativa do pandas para a analises temporais. Este ndice armazena timestamps com preciso de nanossegundos e permite "slicing inteligente". Um utilizador pode solicitar dados com strings parciais, como df['2023-01'], e o ndice resolver automaticamente para selecionar todas as linhas correspondentes a janeiro de 2023, sem a necessidade de operaes de filtragem explcitas complexas.

- **MultiIndex** (Hierarchical Indexing): Permite armazenar dados de dimenses superiores, numa estrutura tabular bidimensional. Atravs de mltiplos nveis de indexao (ex: Ano, depois Trimestre, depois Regio), o pandas oferece uma alternativa flexvel aos painis multidimensionais sem a complexidade de gerenciar estruturas N-dimensionais reais. 




### 2.1.2 *Series*

Uma **Series** (Srie - https://pandas.pydata.org/docs/reference/series.html),  a unidade atmica do pandas,  essncialmente um array **uni**dimensional indexado. Em termos de folha de clculo, representaria uma nica coluna. No entanto, ao contrrio de uma lista simples em Python, uma **Series** possui um ndice (Index)  um conjunto de _rtulos_ intrnsecos aos dados que permite uma recuperao eficiente e alinhamento automtico.   

Se uma lista Python  um "_saco_" de valores acedidos pela sua posio arbitrria (0, 1, 2...), uma **Series** do Pandas  como um dicionrio Python onde as chaves (o ndice) so ordenadas e os valores so "tipados" e homogneos.  

Esta dualidade "array-dicionrio"  o que confere  Series sua flexibilidade:

- Como Array: Suporta slicing (s[0:5]), operaes vetorizadas (s * 2) e funes universais do NumPy (np.log(s)).

- Como Dicionrio: Permite recuperao de valores por chave especifica (s['indice_a']).

Caractersticas das **Series**:

- **Homogeneidade de tipos**: Uma Series armazena tipicamente dados de um nico tipo (inteiro, float, string/objeto). Isto permite as chamadas operaes vetorizadas  a aplicao de uma funo a toda a coluna simultaneamente sem a necessidade de escrever um ciclo for explcito, resultando numa performance computacional superior. Anlogo ao que fazemos hoje no Excel ou em SQL.

- **Imutabilidade de tamanho**: Embora os valores individuais possam ser alterados, o tamanho de uma Series no pode ser modificado "in-place" sem criar um novo objeto. Isto significa que posso editar os valores existentes, mas no consigo adicionar novos valores a essa Serie, e se necessito de o fazer, tenho de criar uma nova Serie com base na original, adicionando os novos valores (explicado no **Anexo A**). Esta caracterstica deriva da sua base em arrays NumPy.   

- **Alinhamento de dados**: As operaes entre Series alinham-se pelo ndice, no pela posio (tambm demonstrado no **Anexo A**). Se somarmos duas Series, o Pandas far a correspondncia dos rtulos do ndice, no dos nmeros das linhas. Isto previne erros comuns no Excel onde o desalinhamento de clulas leva a clculos errados "ocultos". 

- Pode ser criada a partir de listas ou dicionrios.

### 2.1.3 _DataFrame_

O **DataFrame** (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame)  a estrutura de dados principal do Pandas.  uma estrutura tabular **bi**dimensional, mutvel em tamanho, e potencialmente heterognea, com eixos indexados (linhas e colunas).  uma coleo de Series que partilham o mesmo ndice de linhas, no entanto, a sua implementao fsica no BlockManager garante que ele  mais eficiente do que um simples dicionrio de Series.   

**Bi**dimensional, pois o DataFrame possui dois eixos:   

- Eixo 0 (Index): O eixo vertical, representando as linhas.
- Eixo 1 (Columns): O eixo horizontal, representando as colunas.

![alt text](image-1.png)
(fonte: https://github.com/jorisvandenbossche/pandas-tutorial/blob/master/02%20-%20Data%20structures.ipynb)

A maioria das operaes de agregao ou estatstica exige a especificao do eixo. Por exemplo, o mtodo *.mean(axis=0)* calcula a mdia de cada coluna (agregando o valor das linhas), enquanto *.mean(axis=1)* calcular a mdia de cada linha (agregando atravs das colunas).

Caractersticas dos **DataFrame**:

-  a estrutura mais utilizada. 
-  bidimensional (semelhante a uma tabela inteira ou folha de clculo).
-  Possui ndices para as linhas e cabealhos para as colunas:
- Colunas: Cada coluna num DataFrame , na realidade, uma Series.
    - Linhas: As linhas partilham um rtulo de ndice.
    - Clulas: A interseo de uma linha e uma coluna.  
- Pode ser criado a partir de dicionrios, listas ou leitura de ficheiros externos.

Distinguir os dois conceitos **Series**/**DataFrame**,  fundamental. 

Enquanto uma coluna especfica (uma Series) deva ser consistente no seu tipo de dados (por exemplo, todos os valores so inteiros), o DataFrame como um todo pode conter colunas de inteiros ao lado de colunas de texto e colunas de datas.

## 4.2 Ingesto de Dados (ETL: Extraction)

O Pandas oferece uma suite de funes de leitura, funo read_csv (read_csv, read_excel, read_sql, read_json, etc.). Utilizaremos read_csv para carregar os dados diretamente de um repositrio Kaggle.

Teoria por trs do read_csv: Quando esta funo  chamada, o Pandas realiza um varrimento inicial (buffer) para determinar o delimitador (neste caso a virgula) e os tipos de dados de cada coluna. Se uma coluna contiver misturas de nmeros e strings, o Pandas adota o tipo da coluna para o tipo mais genrico (object), o que consome mais memria. ## 4.4 Limpeza e Tratamento de Dados (Data Wrangling)

A limpeza, tratamento e enriquecimento dos dados, ou Data Wrangling (fonte: https://www.ibm.com/think/topics/data-wrangling), consome a maior parte do tempo de um cientista de dados, e o pandas fornece ferramentas bastante eficazes para essa tarefa.

A **limpeza de dados**  guiada por decises sobre a natureza da ausncia dos dados. Conforme vimos atravs da funo `info()`, temos colunas com valores nulos.  Vamos analisar as **Estratgias de Resoluo** para valores nulos:

- **Eliminao** (Dropping): `titanic.dropna(subset=['Age'])` remove as linhas onde a idade  desconhecida. Embora simples, isto reduz o tamanho da amostra e pode introduzir vis se os dados em falta no forem aleatrios.

- **Imputao** (Imputation): Preencher os valores em falta usando `titanic.fillna()`.

    - Imputao **Simples**: Preencher com a idade mdia global (`titanic['Age'].mean()`).

    - Imputao **Contextual**: Preencher com a mediana da idade do gnero e classe especficos. Por exemplo, preencher a idade em falta de um homem da 3 classe com a mediana das idades de outros homens da 3 classe. Esta abordagem preserva a distribuio estatstica e a integridade dos dados muito melhor do que uma mdia global.

No nosso dataframe:

- **Coluna Cabin**: Com mais de 75% dos dados ausentes, a imputao (atribuir um valor) introduziria um desvio estatstico significativo. A deciso mais conservadora, ser a remoo da caracterstica (_Feature Dropping_).

- **Coluna Age**: A idade  um valor preditivo geralmente de grande valor para sobrevivncia. A remoo das linhas reduziria significativamente o tamanho da amostra (perda de informao). A imputao  prefervel. Usaremos a mediana em vez da mdia, pois a mdia  sensvel a outliers (passageiros muito idosos), enquanto a mediana  uma medida de tendncia central com maior significado.   

- **Coluna Embarked**: Possui apenas 2 valores nulos. Remover essas duas linhas (observaes) tem impacto estatstico insignificante, pelo que faremos o drop mas apenas dessas linhas.

```python
# Vamos criar o Pipeline de Limpeza

# Criar uma cpia para preservar o raw data (boa prtica)
titanic_clean = titanic.copy()

# 1. Remoo de coluna com excesso de nulos: Coluna Cabine
titanic_clean.drop(columns=['Cabin'], axis=1, inplace=True)  # axis=1 indica operao na coluna vertical

# 2. Imputao da Idade com a Mediana das idades
mediana_idade = titanic_clean['Age'].median()
titanic_clean['Age'].fillna(mediana_idade, inplace=True)

# 3. Remoo de linhas Embarked com valor nulo
titanic_clean.dropna(subset=['Embarked'], inplace=True)  # axis=0 (padro) indica operao na linha

# Verificao do nosso datafram aps limpeza
print("Valores nulos restantes no dataset limpo:", titanic_clean.isnull().sum().max())
print("------------------------------------")
```

```python
titanic_clean.info()
```

Verificamos que a coluna `Age` j no tem valores nulos! 

Verificamos que j no existe a coluna `Cabin`! 

E Verificamos que temos no total 889 registos, pois 2 foram eliminados devido ao valor nulo da coluna `Embarked`.

**Feature engeneering (ou engenharia de variaveis)**: O Feature Engineering, ou engenharia de variveis  o processo de usar conhecimento de domnio para criar novas variveis com valor para a analise ou pipeline de dados.. Pode ser por exemplo pela 

- **Manipulao de Texto**, .str expe mtodos vetorizados para manipulao de strings (ex: df['nome'].str.upper(), df['email'].str.contains('@')). Isso permite aplicar operaes de string a colunas inteiras sem a penalidade de performance de loops explcitos em Python. 

- Pode ser atravs de **Colunas calculadas**, onde com base em outras colunas j existente criamos uma nova coluna.

#### Manipulao de Texto

Vamos criar campos separados para UltNome (sobrenome) e Nome (restantes nomes) para anlise de famlias e normalizao de nomes, e poder enriquecer os nossos dados com novas "features".

```python
titanic_clean['UltNome'] = titanic_clean['Name'].str.split(',').str[0].str.strip()         # Como os nomes esto no padro Sobrenome, Ttulo. PrimeiroNome. Para UltNome ficamos com o que est antes e da virgula
titanic_clean['Nome'] = titanic_clean['Name'].str.split(',').str[1].str.strip()            # Neste caso ficamos com o que est aps a virgula.
titanic_clean[['Name','UltNome','Nome']].head(5)                                     # Vamos ver as primeiras 5 linhas para verificar as duas novas colunas.
```

Outro exemplo ser detetar possveis famlias por sobrenome repetido, e para isso criamos uma nova "feature" com um booleano, se o nome  repetido ou no.

```python
UltNome_counts = titanic_clean['UltNome'].value_counts()
titanic_clean['NomeFamiliaRepetido?'] = titanic_clean['UltNome'].map(lambda s: UltNome_counts.get(s, 0) > 1)
titanic_clean[['Name','UltNome','NomeFamiliaRepetido?']].head(8)
```

Podemos, tambm com recurso  extrao de texto, relacionar o titulo (presente no nome) com a taxa de sobrevivncia.

```python
# Usando **titanic_clean['Name'].str.extract(' ([A-Za-z]+)\.')**, esta extrao baseada em expresses regulares (_RegEx_) permite-nos agrupar no apenas por gnero, mas por ttulo social. 
titanic_clean['Title'] = titanic_clean['Name'].str.extract(r' ([A-Za-z]+)\.')

# Calculamos a media de sobrevivncia por Titulo
titanic_clean.groupby('Title')['Survived'].mean() 
    
```

Ficamos a saber, por exemplo, que "Master" (um ttulo vitoriano para jovens rapazes) tem uma taxa de sobrevivncia muito superior a "Mr." (homens adultos), e parece tambm confirmando a aplicao estrita da "regra social" ***mulheres e crianas primeiro***.   

### Colunas calculadas

Vamos criar uma nova coluna `TamanhoFamilia` somando SibSp (irmos/cnjuges) e Parch (pais/filhos).


```python
titanic_clean['TamanhoFamilia'] = titanic_clean['SibSp'] + titanic_clean['Parch']     #Somamos duas colunas para originar a nova coluna
display(titanic_clean[['Name', 'TamanhoFamilia']].head(3))                #Vamos ver as primeiras 3 linhas para verificar a nova coluna, calculada.
```

Agora que temos a nova coluna (feature) "TamanhoFamilia" j podemos relacionar com a coluna "Survive", calculando a mdia por cada "tamanho":

```python
# RELACIONAR familysize com sobrevivencia
sobrevivencia_por_familia = titanic_clean.groupby('TamanhoFamilia')['Survived'].mean()
sobrevivencia_por_familia
```

Concluso possvel destes novos dados, seria que as maiores taxas de sobrevivncia foram para agregados de 3 ou menos pessoa, ou seja, que famlias com mais que 3 elementos dificilmente sobreviveram. ## 4.4  Indexao, Seleo e Filtragem

Uma vez carregados os dados, feita uma analise inicial, limpeza e engenharia de variveis, outra tarefa comum  selecionar subconjuntos de interesse, para responder a diferentes questes. O pandas oferece mltiplos paradigmas de seleo, o que historicamente gerou alguma confuso, e levou  padronizao atual fas funes **.loc** e **.iloc**.

Um dos obstculos mais comuns quando nos iniciamos no usa da biblioteca  a confuso entre selecionar dados por rtulo (label) versus selecionar dados por posio (Integer Location), que so utilizados respetivamente pelos comandos **.loc** e **.iloc**.  

- **Seleo Baseada em Rtulos (.loc)**:  a forma mais explcita e recomendada. `df.loc['linha_a', 'coluna_b']` recupera dados baseados estritamente nos nomes do ndice e das colunas. Uma paticularidade do slicing neste caso,  que  inclusiva em ambas as extremidades (ao contrrio da regra padro do Python).

- **Seleo Baseada em Posio (.iloc)**: Funciona como a indexao de listas ou arrays NumPy. `df.iloc[0:5, 1]` seleciona as primeiras cinco linhas da segunda coluna. Aqui, o slicing segue a regra padro do Python (exclui o ltimo valor, neste caso o 5).

- **Indexao Booleana**:  a ferramenta essencial para filtragem exploratria. Expresses como df[df['Idade'] > 18] criam uma mscara booleana (uma Series de True/False) que  usada para filtrar o DataFrame, retornando apenas as linhas onde a condio  verdadeira. Permite ainda o uso de operadores (&, |, ~) para combinar mltiplas condies.

- **Filtros**: Uma das grandes funcionalidaeds do pandas,  a filtragem de dados baseada em condies lgicas. Isto utiliza a lgica vetorizada, permitindo perguntas complexas aos dados.

#### **iloc** (Localizao por Posio)

O indexador .iloc  puramente baseado em posies inteiras (de 0 a comprimento-1). O limite superior com iloc  exclusivo (o ndice final no  includo), mantendo a consistncia com a indexao do Python

- Sintaxe: df.iloc[ndice_linha, ndice_coluna]

Exemplo: titanic.iloc retorna os valores da primeira linha do conjunto de dados Titanic (Sr. Owen Harris Braund).   

Exemplo de aplicao a um intervalo: titanic.iloc[0:5, 0:2] retorna as primeiras 5 linhas e as primeiras 2 colunas,  que so "PassangerId" e "Survived".

```python
titanic_clean.iloc[0]  # Primeira linha do DataFrame
```

```python
titanic_clean.iloc[0:5, 0:2]  # Primeiras 5 linhas e primeiras 2 colunas do DataFrame
```

#### **loc**: Label Location (Localizao por _Rtulo_)

O indexador .loc  baseado em rtulos.

- Sintaxe: df.loc[rtulo_linha, nome_coluna]

- Exemplo: df.loc[0, 'Name'] retorna o nome associado ao rtulo de ndice 0.

O limite superior no fatiamento com loc  inclusivo. Esta  a maior diferena com o iloc: titanic.loc[0:5] retorna 6 linhas (os rtulos 0, 1, 2, 3, 4 e 5). 

A razo para isto  que, se os rtulos fossem strings (ex: "Jan", "Fev", "Mar"), no seria intuitivo saber qual  o "prximo" elemento para fazer a excluso, pelo que o Pandas opta por incluir o ltimo rtulo explicitamente.

No inicio da utilizao do pandas, esta grande diferena gera confuso, porque o ndice padro do Titanic parece ser nmeros inteiros (0, 1, 2...). Assim, titanic.loc e titanic.iloc parecem fazer a mesma coisa. No entanto, se ordenssemos o DataFrame por Idade (titanic.sort_values('Age')), a linha na posio 0 (a primeira linha fsica) poderia ter o rtulo 630. Nesse caso:
- **iloc** retornaria o passageiro mais jovem (que est na primeira posio fsica).
- **loc** retornaria o passageiro com o ID 0 (que agora pode estar no meio da tabela).

```python
titanic_clean.loc[0:5]  # Primeiras 6 linhas do DataFrame usando loc
```

## 4.5 Agregao

### **Group by**

O mtodo **groupby**  uma funo fundamental na anlise exploratria. Ele segue um conceito conhecido como **Split-Apply-Combine** (Dividir-Aplicar-Combinar).   

- **Split** (Dividir): Os dados so particionados em grupos com base em chaves (ex: dividir os passageiros por Pclass).

- **Apply** (Aplicar): Uma funo  aplicada a cada grupo independentemente (ex: calcular a mdia de Survived).

- **Combine** (Combinar): Os resultados so fundidos novamente numa nica estrutura de dados.

Relembrar: Eixos (0 e 1)

Muitas funes (como drop, mean, sum) exigem um argumento axis (eixo).

- **Axis 0** (ndice/Linhas): As operaes movem-se para baixo ao longo do eixo vertical. 
   - titanic.mean(axis=0) calcula a mdia de cada coluna. 
   -  a direo do ndice.

- **Axis 1** (Colunas): As operaes movem-se atravs do eixo horizontal. 
    - titanic.drop('col_name', axis=1) remove uma coluna.   

Com o `group by`, de forma simples e rpida conseguimos entender correlaes, como por exemplo a sobrevivncia por classe.

```python
sobrevivencia_classe = titanic_clean.groupby('Pclass')[['Survived']].mean()
display(sobrevivencia_classe)
```

Ao analisarmos o output, descobrimos uma estratificao clara:

- Classe 1: ~63% de sobrevivncia

- Classe 2: ~47% de sobrevivncia

- Classe 3: ~24% de sobrevivncia 

Esta simples linha de cdigo Pandas revela a desigualdade, que podemos classificar de social, confirmando que o estatuto socioeconmico foi um determinante na probabilidade de vida ou morte.

Podemos usar o `group by` associando o `.mean`, e saber por exemplo a mdia de idades dos passageiros, por gnero.

```python
print("--- Agrupando por Sexo e Calculando a Mdia de Idade ---")
titanic_clean.groupby('Sex')['Age'].mean()
```

#### **Agrupamento Multi-Nvel**

Podemos agrupar por mltiplas colunas para obter outputs mais complexos.

Por exemplo agrupar por Classe e Genero, isto vai revelar que, embora os passageiros da 1 Classe tenham sobrevivido a taxas elevadas, as Mulheres da 1 Classe tiveram uma taxa de sobrevivncia de quase 97%, enquanto os Homens da 3 Classe tiveram a mais baixa de todas. 

Este cruzamento de variveis demonstra como o Pandas facilita a anlise multivariavel.

#### **.agg()**

O mtodo **.agg()** permite aplicar mltiplas funes estatsticas simultaneamente. 

Por exemplo, conseguimos ver no apenas a taxa de sobrevivncia, mas tambm o nmero (count) de passageiros em cada classe, fornecendo contexto estatstico (tamanho da amostra) para as taxas calculadas. 


```python
# Do dataframe, vamos usar a coluna Pclass e Survived, e agrupar por Pclass, calculando a mdia e o count de Survived
titanic_clean[['Pclass', 'Survived']].groupby('Pclass').agg(['mean', 'count'])
```

A anlise revela uma relao (no linear), onde parece que estar sozinho (FamilySize=0) tem baixa sobrevivncia, ter uma famlia pequena (1-3) aumenta a sobrevivncia, mas famlias grandes (>4) tm taxas de sobrevivncia muito baixas, provavelmente devido  dificuldade logstica de reunir todos os membros durante a evacuao catica.

## Anexo A

- Como mudar um valor de uma Series (funciona in-place).
- Como tentar mudar o tamanho e perceber que isso gera uma nova Series (novo objeto).

```python
import pandas as pd
import numpy as np

# 1) Criar uma Series. Vemos claramente que permite uma logica similar a um dicionrio em Python, ao ter a possibilidade de definir o indice de cada elemento da serie. Se no definesse atribuia automaticamente.
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("Series original:\n", s) # imprimimos a serie que acabamos de criar para confirmar a sua composio.

# 2) Agora pretendemos mostrar que podemos mudar um valor (isto altera 'in-place')
s['b'] = 99  #usando o tal indice que existem sempre numa serie, vamos mudar o indice "b" para 99.
print("\nDepois de alterar o valor em 'b':\n", s) # imprimimos a serie novamente para visualizar que assume o novo valor para o indice "b".

# 3) Tentar "mudar o tamanho" usando operaes que NO modificam o objeto original. Ex.: tentamos um drop, e este retorna uma NOVA Series, e se no guardar o resultado, "s" continua igual
s_sem_a = s.drop('a')  # cria uma nova Series em que fazemos drop do indice "a"
print("\nResultado de drop('a') (nova Series):\n", s_sem_a)
print("\nA Series original continua igual:\n", s)

# 4) Para de facto mudar o tamanho (adicionar/remover), precisa criar/receber um NOVO objeto Ex.: adicionar um elemento via CONCAT (gera uma nova Series)
novo_elemento = pd.Series([40], index=['d'])
s_expandida = pd.concat([s, novo_elemento])
print("\nSeries expandida (nova Series):\n", s_expandida)

# Se quiser que 's' fique com o novo tamanho, tem de reatribuir a nova serie  variavel "s"
s = s_expandida
print("\nReatribumos a 's' para ficar com o novo tamanho:\n", s)

# 5) Exemplo relacionado com NumPy (a base de uma Series):
arr = np.array([1, 2, 3])
print("\nNumPy array original:", arr)

# Alterar um valor  in-place
arr[1] = 99
print("NumPy array aps alterar valor:", arr)

# "Redimensionar" o array no  in-place; np.append cria um NOVO array
arr2 = np.append(arr, 4)
print("Novo NumPy array aps append (arr2):", arr2)
print("O array original continua:", arr)
```

## Anexo B

```python
# MERGE

# DataFrame A
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'nome': ['Ana', 'Bruno', 'Carlos']
})

# DataFrame B
df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'cidade': ['Lisboa', 'Porto', 'Coimbra']
})

# Merge INNER (apenas ids comuns)
inner_merge = pd.merge(df1, df2, on='id', how='inner')

# Merge LEFT (todos do df1)
left_merge = pd.merge(df1, df2, on='id', how='left')

# Merge OUTER (todos os ids)
outer_merge = pd.merge(df1, df2, on='id', how='outer')

print(inner_merge)
print(left_merge)
print(outer_merge)
```

```python
# CONCAT

# DataFrames com mesmo ndice
df3 = pd.DataFrame({'idade': [25, 30, 35]}, index=['Ana', 'Bruno', 'Carlos'])
df4 = pd.DataFrame({'cidade': ['Lisboa', 'Porto', 'Coimbra']}, index=['Ana', 'Bruno', 'Carlos'])

# Concat horizontal (axis=1)
concat_horizontal = pd.concat([df3, df4], axis=1)

# Concat vertical (axis=0)
concat_vertical = pd.concat([df3, df3], axis=0)

print(concat_horizontal)
print(concat_vertical)
```
